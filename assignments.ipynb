```python
def new_zeros(C, rows, cols):
    alpha = C.drop(cols, axis=1).drop(rows).min().min()
    C = C.copy()
    for i, j in itertools.product(C.index, C.columns):
        times = (i in rows) + (j in cols)
        if times == 0:
            C.loc[i,j] -= alpha
        elif times == 2:
            C.loc[i, j] += alpha
    return C            
```
